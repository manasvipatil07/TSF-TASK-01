<h2>TASK 01 : Prediction using Supervised ML</h2> 

<h3>Importing the libraries</h3>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

<h3>Reading the data from the provided link</h3>

```python
data=pd.read_csv("http://bit.ly/w-data")
data.head(8)
```

<h3>Check any missing values</h3>

```python
 data.isnull().sum()
```

```python
plt.scatter(x=data.Hours,y=data.Scores,color="maroon")
plt.title("Scatter Plot")
```

<h3>Splitting the data</h3>

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

<h3>Splitting the data into training  & testing sets</h3>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h3>Linear Regression Model</h3>

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

<h3>Data Visualization</h3>
<h4>Scatter Plot</h4>

```python
# Scatter plot for training set
plt.scatter(X_train, y_train, color='maroon')
plt.plot(X_train, regressor.predict(X_train), color='indigo')
plt.title('Hours vs Scores (Training set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()

# Scatter plot for testing set
plt.scatter(X_test, y_test, color='maroon')
plt.plot(X_train, regressor.predict(X_train), color='indigo')
plt.title('Hours vs Scores (Testing set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

<h4>Bar Plot</h4>

```python
plt.figure(figsize=(8, 6))

# Bar plot for training set
plt.subplot(2, 1, 1)
plt.bar(X_train.flatten(), y_train, color='#A9DFBF', label='Actual Scores')
plt.plot(X_train.flatten(), regressor.predict(X_train), color='maroon', label='Predicted Scores', linestyle='--')
plt.title('Hours vs Scores (Training set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()

# Bar plot for testing set
plt.subplot(2, 1, 2)
plt.bar(X_test.flatten(), y_test, color='#A9DFBF', label='Actual Scores')
plt.plot(X_train.flatten(), regressor.predict(X_train), color='maroon', label='Predicted Scores', linestyle='--')
plt.title('Hours vs Scores (Testing set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Number of hours for prediction
hours_to_predict = np.array([[9.25]])
# Predicting the score
predicted_score = regressor.predict(hours_to_predict)
print("Predicted score if a student studies for 9.25 hours/day:", predicted_score[0])
```
